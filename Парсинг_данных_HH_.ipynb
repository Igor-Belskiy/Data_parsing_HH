# Обрабатываем информацию о поле и возрасте соискателей

```
# Функция для данных о поле

def getParameterSex(arg):

  if "Муж" in arg:
    sex='М'
  elif "Жен" in arg:
    sex = 'Ж'
  else:
    sex = "Не указан"
  return sex


# Функия для данных о возрасте

def getParameterAge(arg):
  out = 0
  year_tec = 2021               # задаётся текущий год 
  if (len(arg) > 7):
    out = year_tec - int(re.findall(r'\d{4}', arg)[0])

  return str(out)
```

```
# Вставляем столбец "Пол" и заполняем его с помощью функции

df.insert(0,'Пол', np.array( [getParameterSex(df['Пол, возраст'][i]) for i in range(len(df))] ) )   
```

```
# Вставляем столбец "Возраст" и заполняем его с помощью специальной функции

df.insert(1,'Возраст', np.array( [getParameterAge(df['Пол, возраст'][i]) for i in range(len(df))] ) )  
```

```
# После обработки удаляем исходный столбые "Пол, возраст"

df = df.drop('Пол, возраст', axis = 1)
```

# Обрабатываем столбец "Зарплата"

```
#  Находим все виды валют, которые встречаются в столбце зарплата

print( pd.Series( [ re.search(r'[a-zA-Zа-яА-Я]+', df['ЗП'][i])[0]   for i in range (df.shape[0]) ] ).unique() )
  
```

```
# Функция обработки поля "Зарплата"

def getParameterSalary(arg):
    
    # Сначала получаем чистое число, убираем лишние знаки
    salary = 0
    arg=str(arg)

    salary = float( re.search(r'[0-9]+', arg)[0] )
    curr = re.search(r'[a-zA-Zа-яА-Я]+', arg)[0]

    # Применяем курсы валют приводя тем самым итоговую зарплату к рублям.

    if 'USD' in curr :   salary = salary * 74.6     # Доллар
    elif 'KZT' in curr :  salary = salary * 0.18    # Казахский тенге
    elif 'EUR' in curr :  salary = salary * 89.83   # Евро
    elif 'AZN' in curr :  salary = salary * 44.6    # Азербайджанский манат
    elif 'сум' in curr :  salary = salary * 0.0071  # Узбекский сум
    elif ('грн' in curr) or ('UAH' in curr) :  salary = salary * 2.7     # Украинская гривна
    elif ('бел' in curr) or ('BYN' in curr) :  salary = salary * 29.24   # Белорусский рубль
    elif ('KGS' in curr) or ('som' in curr) :  salary = salary * 0.88    # Киргизский сом
 
    return str(int(salary))
```

```
# Подготовленной функцией обрабатывем столбец с зарплатой

for i in range(df.shape[0]) :
  df['ЗП'][i] = getParameterSalary( df['ЗП'][i] )
```

```
#Изменяем названия столбца: "ЗП" на "Зарплата (руб)".

df = df.rename(columns={'ЗП':'Зарплата (руб)'})

df.head()
```

```
#Функция обработки данных о готовности к переезду

def getParameterRelocate(a):
  
  if (('перее' in a) or ('relocate' in a)) and not (('не' in a) or ('not' in a)) : return 'готов к переезду'

  else : return 'не готов к переезду'
```

# Обрабатываем данные о городе

```
# Функция обработки данных о городе

def getParameterCity(arg):
  millionCities = "Новосибирск Екатеринбург Нижний Новгород Казань Челябинск Омск Самара Ростов-на-Дону Уфа Красноярск Пермь Воронеж Волгоград"
  sarg=arg.split(',')
  item = sarg[0]     # Берем только город
  item= item.strip() # Удаляем пробелы
  if item == "Москва":          return "Москва"
  if item == "Санкт-Петербург": return "Санкт-Петербург"
  if item in millionCities:     return item
  else: return "малый или за пределами России город"
```

```
# Фунция обработки данных о желаемой занятости 
# Приводим к единообразному набору терминов в базе по теме занятости

def getParameterEmployment(arg):
  
  result = ''

  if ("стажировка" in arg)|("volunteering" in arg)|("work placement" in arg)|("волонтерство" in arg): result += 'стажировка, '
  if ("частичная занятость" in arg)|("part" in arg):  result += 'частичная занятость, '
  if ("проектная работа" in arg)|("project" in arg):  result += 'проектная работа, '

  if ("полная занятость" in arg)|("full" in arg):     result += 'полная занятость'
  elif result != '' : result = result[0:len(result)-2]                               # В этой строке удаляем ненужные запятую и пробел в хвосте, если они образовались

  if result == '': result='Не указано'

  return result
```

```
# Функция обработки столбца "График"

def getParameterSchedule(arg):

  result = ''
  if ("гибкий график" in arg)|("flexible" in arg):    result += 'гибкий график, '
  if ("полный день" in arg)|("full" in arg):     result += 'полный день, '
  if ("сменный график" in arg)|("shift" in arg):   result += 'сменный график, '
  if ("вахтовый метод" in arg)|("rotation" in arg):   result += 'вахтовый метод, '

  if ("удаленная работа" in arg)|("remote" in arg): result += 'удаленная работа'
  elif result != '' : result = result[0:len(result)-2]                           # В этой строке удаляем ненужные запятую и пробел в хвосте, если они образовались

  if result=='':  result='Не указано'
  return result
```

```
# Подготовленной функцией обрабатывем столбец "График"

for i in range(df.shape[0]) :
  df['График'][i] = getParameterSchedule( df['График'][i] )
```

```
# Данные об образовании
def getParameterEducation(arg):

  result = ''
  arg = arg.lower()

  if ("высшее" in arg)|("higher education"  in arg): result += 'высшее, '
  if ("среднее специальное"  in arg) | ("special"  in arg) | ("колледж"  in arg) | ("college"  in arg) : result += 'среднее специальное, '
  if ("неоконченное высшее"  in arg) | ("incomplete"  in arg): result += 'неоконченное высшее, '

  if ("среднее образование"  in arg) | ("secondary education"  in arg): result += 'среднее образование'
  elif result != '' : result = result[0:len(result)-2]                           # В этой строке удаляем ненужные запятую и пробел в хвосте, если они образовались


  if result=='':    result='Не указано'

  return result
```

```
# Подготовленной функцией обрабатывем столбец "Образование"

for i in range(df.shape[0]) :
  df['Образование'][i] = getParameterEducation( df['Образование'][i] )
```

```
# Функция обработки столбца "Опыт работы (мес)"

def getParameterExperience(arg):
 
  cifr_list = re.findall(r'\d+',arg)

  # Если в строке найдено 2 группы цифр, тогда первую считаем годом, а вторую месяцем.
  if len(cifr_list) == 2  : result = int(cifr_list[0])*12 + int(cifr_list[1])    

  #Если нашли только одну цифру, проверяем это было количество лет опыта или количество месяцев
  elif (len(cifr_list) == 1) and ( ('year' in arg) or ('лет' in arg) or ('год' in arg) ) :   result = int(cifr_list[0]) * 12   
  elif (len(cifr_list) == 1) and ( ('month' in arg) or ('месяц' in arg) )                :   result = int(cifr_list[0])

  else : result = 0

  return str(result)
```

```
# Подготовленной функцией обрабатывем столбец "Опыт работы (мес)"

for i in range(df.shape[0]) :
  df['Опыт работы (мес)'][i] = getParameterExperience( df['Опыт работы (мес)'][i] )
```
